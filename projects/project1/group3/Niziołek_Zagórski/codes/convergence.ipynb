```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
subplot_index = 1
plt.figure(figsize=(10, 10))

for dataset_type in ["small", "large"]:
    dataset_count = 3 if dataset_type == "small" else 6
    for dataset_index in range(dataset_count):
        plt.subplot(3, 3, subplot_index)
        subplot_index += 1

        dataset_name = f"{dataset_type}{dataset_index+1}"
        df = pd.read_csv(f"convergence/{dataset_name}.csv")
        df = pd.melt(df, id_vars=["Iteration"], var_name="algorithm")

        sns.lineplot(
            data=df,
            x="Iteration",
            y="value",
            hue="algorithm",
            legend=subplot_index == 2,
        ).set(
            title=dataset_name,
            xlabel=None,
            ylabel=None,
        )

plt.tight_layout()

plt.savefig(
    f"plots/convergence.eps",
    # bbox_inches="tight",
)
plt.show()
```
